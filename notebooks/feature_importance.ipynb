```python
import shap
import matplotlib.pyplot as plt
import joblib
import warnings

warnings.filterwarnings("ignore")
```

```python
model_path = '../../ml_project/models/xgb_final.joblib'

def load_model():
    model = joblib.load(model_path)
    return model
model = load_model()
model = model.named_steps['classifier']
```

```python
import pandas as pd
```

```python
X = pd.read_csv('../../ml_project/data/train_data.csv')
X.drop('Class', axis=1, inplace=True)
X = X.sample(30000)
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer(X)
```

```python
shap.summary_plot(shap_values, X,
                  plot_type="bar",
                  max_display=20,
                  show=False)
plt.tight_layout()
plt.show()
```

```python
shap.summary_plot(shap_values, X,
                  max_display=15,
                  plot_size=(10, 7),
                  show=False)
plt.show()
```
